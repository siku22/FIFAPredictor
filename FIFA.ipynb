```python
def predict_match_by_name(team_a_name, team_b_name):
    team_a = team_features[team_features["team"] == team_a_name].iloc[0]
    team_b = team_features[team_features["team"] == team_b_name].iloc[0]

    input_data = pd.DataFrame({
        "overall_diff": [team_a["avg_overall"] - team_b["avg_overall"]],
        "age_diff": [team_a["avg_age"] - team_b["avg_age"]],
        "balance_diff": [team_a["team_balance"] - team_b["team_balance"]],
        "player_pool_diff": [team_a["num_players"] - team_b["num_players"]],
        "rank_advantage": [team_b["rank"] - team_a["rank"]],
        "points_diff": [team_a["total_points"] - team_b["total_points"]],
    })

    prob = model.predict_proba(input_data)[0][1]

    return {
        "team_a": team_a_name,
        "team_b": team_b_name,
        "team_a_win_prob": prob,
        "team_b_win_prob": 1 - prob,
        "winner": team_a_name if prob > 0.5 else team_b_name
    }
```

```python
def predict_match_by_name(team_a_name, team_b_name):
    tf = team_features.copy()
    tf["team_clean"] = tf["team"].str.lower().str.strip()

    team_a_name = team_a_name.lower().strip()
    team_b_name = team_b_name.lower().strip()

    team_a_row = tf[tf["team_clean"] == team_a_name]
    team_b_row = tf[tf["team_clean"] == team_b_name]

    if team_a_row.empty:
        raise ValueError(f"Team A not found: {team_a_name}")
    if team_b_row.empty:
        raise ValueError(f"Team B not found: {team_b_name}")

    team_a = team_a_row.iloc[0]
    team_b = team_b_row.iloc[0]

    input_data = pd.DataFrame({
        "overall_diff": [team_a["avg_overall"] - team_b["avg_overall"]],
        "age_diff": [team_a["avg_age"] - team_b["avg_age"]],
        "balance_diff": [team_a["team_balance"] - team_b["team_balance"]],
        "player_pool_diff": [team_a["num_players"] - team_b["num_players"]],
        "rank_advantage": [team_b["rank"] - team_a["rank"]],
        "points_diff": [team_a["total_points"] - team_b["total_points"]],
    })

    prob_a = model.predict_proba(input_data)[0][1]
    prob_b = 1 - prob_a

    winner = team_a["team"] if prob_a > prob_b else team_b["team"]

    return {
        "team_a": team_a["team"],
        "team_b": team_b["team"],
        "team_a_win_prob": prob_a,
        "team_b_win_prob": prob_b,
        "winner": winner
    }
```

```python
import numpy as np

def simulate_match(team_a, team_b):
    result = predict_match_by_name(team_a, team_b)
    prob = result["team_a_win_prob"]

    return team_a if np.random.rand() < prob else team_b
```

```python
teams = ["brazil", "france", "argentina", "england",
         "spain", "germany", "portugal", "netherlands"]

def simulate_tournament(teams):
    current = teams.copy()

    while len(current) > 1:
        next_round = []
        for i in range(0, len(current), 2):
            winner = simulate_match(current[i], current[i+1])
            next_round.append(winner)
        current = next_round

    return current[0]
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

# -------- FAST LOAD (LOCAL FILES) --------
@st.cache_resource
def load_data():
    model = joblib.load("fifa_model.pkl")
    team_features = joblib.load("team_features.pkl")
    return model, team_features

model, team_features = load_data()

# -------- PAGE CONFIG --------
st.set_page_config(page_title="2026 FIFA Predictor", page_icon="⚽", layout="wide")

st.title("⚽ 2026 FIFA World Cup Predictor")

st.caption("Predict match outcomes using FIFA rankings, player attributes, and team composition.")

st.divider()

# -------- CLEAN TEAM LIST --------
teams = sorted(team_features["team"].dropna().unique())

# -------- INPUT --------
col1, col2 = st.columns(2)

with col1:
    team_a = st.selectbox("Team A", teams)

with col2:
    team_b = st.selectbox("Team B", teams)

# -------- PREDICTION FUNCTION --------
def predict(team_a_name, team_b_name):
    tf = team_features.copy()
    tf["team_clean"] = tf["team"].str.lower().str.strip()

    a = tf[tf["team_clean"] == team_a_name.lower()].iloc[0]
    b = tf[tf["team_clean"] == team_b_name.lower()].iloc[0]

    input_data = pd.DataFrame({
        "overall_diff": [a["avg_overall"] - b["avg_overall"]],
        "age_diff": [a["avg_age"] - b["avg_age"]],
        "balance_diff": [a["team_balance"] - b["team_balance"]],
        "player_pool_diff": [a["num_players"] - b["num_players"]],
        "rank_advantage": [b["rank"] - a["rank"]],
        "points_diff": [a["total_points"] - b["total_points"]],
    })

    prob = model.predict_proba(input_data)[0][1]
    return prob, 1 - prob

# -------- BUTTON --------
if st.button("Predict", use_container_width=True):

    if team_a == team_b:
        st.warning("Pick two different teams.")
    else:
        prob_a, prob_b = predict(team_a, team_b)
        winner = team_a if prob_a > prob_b else team_b

        st.divider()

        # -------- RESULT --------
        st.subheader(f"🏆 {winner} is more likely to win")

        c1, c2 = st.columns(2)

        with c1:
            st.metric(team_a, f"{prob_a:.2%}")

        with c2:
            st.metric(team_b, f"{prob_b:.2%}")

        # -------- BAR --------
        st.progress(float(prob_a))

        st.success(f"⚽ Prediction favors {winner}")

st.divider()

# -------- MODEL INFO --------
st.caption("Model: Logistic Regression | Accuracy: ~69% | Data: 2005–Present")
```
